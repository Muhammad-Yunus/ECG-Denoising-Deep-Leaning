### Inference Model Denoising & Classification


```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from keras import models
```

- load model

```python
denoising_Deep_AE_model = models.load_model('best_denoising_deep_AE.h5')
denoising_Conv_AE_model = models.load_model('best_denoising_conv_AE.h5')
denoising_LSTM_AE_model = models.load_model('best_denoising_LSTM_AE.h5')

cnn_model_Deep_AE = models.load_model('CNN_Classification_model_Deep-AE.h5')
cnn_model_Conv_AE = models.load_model('CNN_Classification_model_Conv-AE.h5')
cnn_model_LSTM_AE = models.load_model('CNN_Classification_model_LSTM-AE.h5')

```

```python
label_names = ['AF', 'N']
```

```python
dataset_folder ="dataset/"
filenames = []
for filename in os.listdir(dataset_folder):
    if filename.find("AFDB_pasien_") > -1:
        filenames.append(filename)
```

```python
filenames
```

- load inference data

```python
inference_data = {}
for name in filenames:
    infer = pd.read_csv(dataset_folder + name, header=None)
    inference_data[name] = infer
```

```python
inference_data[filenames[0]].head()
```

```python
mat_infer_data = {}
for name in filenames:
    infer_df = inference_data[name]
    mat_infer_data[name] = infer_df.values
```

- Apply denoising model to inference data

```python
Deep_AE_den_res = {}
Conv_AE_den_res = {}
LSTM_AE_den_res = {}
for name in filenames:
    Deep_AE_den_res[name] = denoising_Deep_AE_model.transform(mat_infer_data[name])
    Conv_AE_den_res[name] = denoising_Conv_AE_model.transform(mat_infer_data[name])
    LSTM_AE_den_res[name] = denoising_LSTM_AE_model.transform(mat_infer_data[name])
    print("denoising data - %s" % name)
```

```python
def plot_denoising_test(y, x, model='lms'):
    # show results
    plt.figure(figsize=(6,4))
    plt.title("Denoising Result %s - Inference Dataset" % model.upper())
    plt.xlabel("samples - n")
    plt.plot(y,"b", label="y - denoised signal")
    plt.plot(x,"g", label="x - original signal")
    plt.tight_layout()
    plt.legend()
    plt.grid()
    plt.show()
```

- Check result

```python
name = 'AFDB_pasien_AF_Ibu EEM.csv'
idx = 13

y = Deep_AE_den_res[name][idx]
x = mat_infer_data[name][idx]

plot_denoising_test(y, x, model='Deep AE')
```

- Reshape matrix result of denoising

```python
lms_mat_3d = {}
nlms_mat_3d = {}
rls_mat_3d = {}
for name in filenames:
    lms_mat_3d[name] = Deep_AE_den_res[name].reshape(len(Deep_AE_den_res[name]), Deep_AE_den_res[name].shape[1], 1)
    nlms_mat_3d[name] = Conv_AE_den_res[name].reshape(len(Conv_AE_den_res[name]), Conv_AE_den_res[name].shape[1], 1)
    rls_mat_3d[name] = LSTM_AE_den_res[name].reshape(len(LSTM_AE_den_res[name]), LSTM_AE_den_res[name].shape[1], 1)
```

- Apply to CNN model

```python
cnn_lms_classif_result = {}
cnn_nlms_classif_result = {}
cnn_rls_classif_result = {}
for name in filenames:
    cnn_lms_classif_result[name] = cnn_denoising_Deep_AE_model.predict(lms_mat_3d[name])
    cnn_nlms_classif_result[name] = cnn_ndenoising_Conv_AE_model.predict(nlms_mat_3d[name])
    cnn_rls_classif_result[name] = cnn_denoising_LSTM_AE_model.predict(rls_mat_3d[name])
    print("predicting data - %s" % name)
```

- Check result

```python
def plot_cnn_result(data, label='lms', confidence=0.0):
    confidence = confidence*100
    # show results
    plt.figure(figsize=(6,4))
    plt.title("Classification Result : %s - confidence (%.2f%%)" % (label, confidence))
    plt.xlabel("samples - n")
    plt.plot(data,"g", label="signal")
    plt.tight_layout()
    plt.legend()
    plt.grid()
    plt.show()
```

```python
pred_result = []
for name in filenames :
    cnn_res = cnn_rls_classif_result[name] # ubah ke lms, nlms dan rls
    print("\n\n\nPrediction Result - %s\n" % name)
    for i, res in enumerate(cnn_res):
        predicted_id = np.argmax(res)
        label = label_names[predicted_id]
        confidence = res[predicted_id]
        label = label if confidence > 0.70 else '?' # threshold for confidence level , set to 70%
        pred_result.append([name, i, label, confidence])
        
        print('%s \t index - %d \t %s \t %.2f %%' % (name, i, label, confidence*100))
```

```python
res = pred_result[0]
data = Deep_AE_den_res[name][res[1]]
plot_cnn_result(data, res[2], res[3])
```

```python
res = pred_result[-6]
data = Deep_AE_den_res[name][res[1]]
plot_cnn_result(data, res[2], res[3])
```

```python
pred_result_df = pd.DataFrame(pred_result, columns=['filename', 'idx', 'label', 'confidence'])
```

```python
pred_result_df.head()
```

```python
for name in filenames:
    print(name)
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    for key in count :
        print("%s \t: %d" % (key, count[key]))
    print('\n')
```

```python
def plot_donat(porsion, title, labels, colors=['red', 'green', 'blue']):
    plt.figure(figsize=(10,5))

    circle = plt.Circle( (0,0), 0.7, color='white')
    plt.pie(porsion, 
            labels=labels, 
            colors=colors,
            autopct='%1.1f%%')

    plt.title(title)
    p = plt.gcf()
    p.gca().add_artist(circle)
    plt.show()
```

```python
for name in filenames:
    count = dict(pred_result_df[pred_result_df['filename'] == name]['label'].value_counts())
    plot_donat(count.values(), name, count.keys())
```
